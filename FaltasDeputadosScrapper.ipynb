```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
https://sapl.al.am.leg.br/sistema/relatorios/presenca
```

```python
dic_sessao = {'tipo':['Audiência Pública','Extraordinária','Ordinária','Reunião Preparatória','Reunião Solene','Sessão Especial'],'codigo':[4,2,1,7,3,5]}

tipo_sessao = []
total_sessao = []
name = []
titular = []
ativo = []
qtd = []
porc = []
qtd_dia = []
porc_dia = []

for d in dic_sessao['codigo']:
  cabecalho = {'user-agent':'Mozilla/5.0'}
  response = requests.get('https://sapl.al.am.leg.br/sistema/relatorios/presenca?legislatura=20&sessao_legislativa=15&tipo='+str(d)+'&exibir_ordem_dia=on&exibir_somente_ativo=on&data_inicio_0=&data_inicio_1=&pesquisar=Pesquisar', headers=cabecalho)
  sopa = BeautifulSoup(response.text,'html.parser')
  lista_sessao = sopa.find_all('b')
  lista = sopa.find_all('td')
  cont = 0
  for a in lista:
    if cont == 0:
      tipo_sessao.append(lista_sessao[3].text)
      total_sessao.append(lista_sessao[4].text)
      n_split = lista[cont].text.split('/')
      name_upp = n_split[0].upper()
      name.append(name_upp)
      titular.append(lista[cont+1].text)
      ativo.append(lista[cont+2].text)
      qtd.append(lista[cont+3].text)
      porc.append(lista[cont+4].text)
      qtd_dia.append(lista[cont+5].text)
      porc_dia.append(lista[cont+6].text)
      print(lista_sessao[3].text,lista_sessao[4].text,lista[cont].text, lista[cont+1].text,lista[cont+2].text,lista[cont+3].text,lista[cont+4].text,lista[cont+5].text,lista[cont+6].text)
    if (cont % 7 == 0) and (cont > 0):
      print(cont)
      tipo_sessao.append(lista_sessao[3].text)
      total_sessao.append(lista_sessao[4].text)
      n_split = lista[cont].text.split('/')
      name_upp = n_split[0].upper()
      name.append(name_upp)
      titular.append(lista[cont+1].text)
      ativo.append(lista[cont+2].text)
      qtd.append(lista[cont+3].text)
      porc.append(lista[cont+4].text)
      qtd_dia.append(lista[cont+5].text)
      porc_dia.append(lista[cont+6].text)
      print(lista_sessao[3].text,lista_sessao[4].text,lista[cont].text, lista[cont+1].text,lista[cont+2].text,lista[cont+3].text,lista[cont+4].text,lista[cont+5].text,lista[cont+6].text)
    cont += 1
print(cont)
```

```python
dic_falta = {'tipo_sessao':tipo_sessao,'total_sessao':total_sessao,'nome':name,'titular':titular,'ativo':ativo,'qtd_sessao':qtd,'porc_sessao':porc,'qtd_dia':qtd_dia,'porc_dia':porc_dia}
```
